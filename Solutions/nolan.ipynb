# Christopher Nolan: Il Regista che ha ridefinito il cinema contemporaneo


Christopher Nolan è uno dei registi più influenti della sua generazione. La sua carriera è caratterizzata da una narrazione innovativa, una predilezione per le storie non lineari e un impatto significativo sia sul pubblico che sulla critica. La sua carriera è segnata da una serie di film che hanno ridefinito il linguaggio cinematografico, alternando successi ad alto budget a opere più intime e sperimentali.

L'impatto di Nolan sul cinema è evidente non solo nel successo commerciale delle sue opere, ma anche nell'influenza che ha esercitato su altri registi e sul pubblico. La sua dedizione alla pellicola in un'epoca dominata dal digitale e il suo rifiuto dell'uso eccessivo della CGI lo hanno reso un punto di riferimento per chi cerca un cinema spettacolare ma allo stesso tempo autentico.

Il suo stile distintivo, caratterizzato da una regia precisa, colonne sonore immersive e un uso innovativo del tempo e dello spazio, ha reso Christopher Nolan uno degli autori più rispettati e riconoscibili del panorama cinematografico contemporaneo.


```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import plotly.express as px

# Ottiene la directory di lavoro corrente
script_dir = os.getcwd()

# Cartella dove si trovano i file CSV (modifica con il percorso corretto)
clean_path = os.path.join(script_dir, "Data", "clean_data")

# Caricamento dei dataset con percorsi dinamici
actors = pd.read_csv(os.path.join(clean_path, "actors_cleaned.csv"))
countries = pd.read_csv(os.path.join(clean_path, "countries_cleaned.csv"))
crew = pd.read_csv(os.path.join(clean_path, "crew_cleaned.csv"))
genres = pd.read_csv(os.path.join(clean_path, "genres_cleaned.csv"))
languages = pd.read_csv(os.path.join(clean_path, "languages_cleaned.csv"))
movies = pd.read_csv(os.path.join(clean_path, "movies_cleaned.csv"))
posters = pd.read_csv(os.path.join(clean_path, "posters_cleaned.csv"))
releases = pd.read_csv(os.path.join(clean_path, "releases_cleaned.csv"))
studios = pd.read_csv(os.path.join(clean_path, "studios_cleaned.csv"))
themes = pd.read_csv(os.path.join(clean_path, "themes_cleaned.csv"))
oscar_awards = pd.read_csv(os.path.join(clean_path, "the_oscar_awards_cleaned.csv"))
rotten_tomatoes = pd.read_csv(os.path.join(clean_path, "rotten_tomatoes_reviews_cleaned.csv"))
```

# Gli Inizi e il Primo Successo (1998-2001)

Christopher Nolan debutta nel mondo del cinema con Following (1998), un thriller noir a basso budget girato in bianco e nero con un cast di attori non professionisti. Realizzato con un budget di circa 6.000 dollari, il film è stato girato nei fine settimana nell'arco di un anno, poiché Nolan e il suo team lavoravano con risorse limitate e attrezzature minime. Nonostante queste restrizioni, Following mostra già molti degli elementi che caratterizzeranno il suo stile distintivo: una narrazione non lineare, l'uso della soggettività nella percezione della realtà e una predilezione per atmosfere cupe e misteriose. Grazie alla sua originalità e alla sua costruzione narrativa, Following ottiene il plauso della critica e diventa un biglietto da visita fondamentale per Nolan, permettendogli di attrarre l'attenzione di produttori e studios.


```python
# Trova l'ID del film "Following"
following_movie = movies[movies['name'] == 'Following']
movie_id = following_movie['id'].values[0]

# Filtra le distribuzioni per il film "Following"
following_releases = releases[releases['id'] == movie_id]

# Ottieni l'elenco dei paesi in cui è stato rilasciato "Following"
countries_released_following = following_releases['country'].unique()

# Creazione di un DataFrame per i paesi rilasciati
df_countries_following = pd.DataFrame(countries_released_following, columns=['Country'])

# Creare la mappa interattiva per i paesi di "Following"
fig = px.choropleth(df_countries_following,
                    locations='Country',
                    locationmode='country names',
                    title="Distribuzione di 'Following' nei Paesi")

# Mostrare la mappa
fig.show()
```

L'immagine mostra la distribuzione dei Paesi raggiunti dal primo film di Nolan, evidenziando che la diffusione è stata piuttosto limitata. In particolare, emerge chiaramente che la maggior parte dei Paesi coinvolti appartiene all’America (soprattutto Stati Uniti e Canada) e all’Europa occidentale. Al di fuori di queste aree, la presenza è molto marginale, con pochissime eccezioni, come il Giappone e il Brasile. Questo conferma che il primo film di Nolan ha avuto una distribuzione iniziale piuttosto ristretta, concentrata prevalentemente nei mercati occidentali e nei Paesi anglofoni.

Il successo arriva nel 2000 con Memento, un thriller psicologico raccontato a ritroso che consacra Christopher Nolan come uno dei registi più innovativi della sua generazione. Basato su un racconto scritto dal fratello Jonathan Nolan, l'originalità della struttura narrativa e la profondità tematica del film, che esplora il rapporto tra memoria, identità e percezione della realtà, gli valgono un enorme successo di critica e pubblico. Memento diventa un cult e riceve numerosi riconoscimenti, tra cui una nomination agli Oscar per la miglior sceneggiatura originale e una ai Golden Globe. La sua notorietà porta Nolan all’attenzione di Hollywood, permettendogli di accedere a produzioni con budget più elevati e di avviare una carriera destinata a rivoluzionare il cinema contemporaneo. Dato interessante risulta dunque il raggruppamento dello score dei punteggi ottenuti dai primi due film del regista, che mostra la crescente popolarità guadagnata:

```python
# Filtrare le recensioni per "Following" e "Memento"
df_following_reviews = rotten_tomatoes[rotten_tomatoes['movie_title'] == 'Following']
df_memento_reviews = rotten_tomatoes[rotten_tomatoes['movie_title'] == 'Memento']

# Estrarre gli score delle recensioni
following_scores = df_following_reviews['review_score'].dropna()
memento_scores = df_memento_reviews['review_score'].dropna()

# Creazione dei grafici
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Grafico per "Following"
ax1.hist(following_scores, bins=20, color='skyblue', edgecolor='black')
ax1.set_title('Distribuzione delle recensioni per Following')
ax1.set_xlabel('Score della recensione')
ax1.set_ylabel('Numero di recensioni')

# Grafico per "Memento"
ax2.hist(memento_scores, bins=20, color='lightgreen', edgecolor='black')
ax2.set_title('Distribuzione delle recensioni per Memento')
ax2.set_xlabel('Score della recensione')
ax2.set_ylabel('Numero di recensioni')

# Visualizzare entrambi i grafici
plt.tight_layout()
plt.show()
```

Following mostra una distribuzione piuttosto frammentata, con voti che oscillano tra il 5 e l’8, e solo qualche picco più alto. Questo riflette un’accoglienza moderata, tipica di un'opera prima indipendente che ha colpito positivamente una nicchia di spettatori, ma non ha generato un consenso unanime.
Memento, al contrario, evidenzia una distribuzione decisamente più polarizzata verso l’alto, con moltissime recensioni comprese tra l’8 e il 10. Questo dato conferma la trasformazione di Nolan da regista emergente a talento acclamato a livello internazionale.

```python
# Trovare l'ID del film "Following" (1998) di Christopher Nolan
following_movie = movies[movies['name'] == 'Following']
movie_id_following = following_movie['id'].values[0]

# Trovare l'ID del film "Memento" (2000) di Christopher Nolan
memento_movie = movies[movies['name'] == 'Memento']
movie_id_memento = memento_movie['id'].values[0]

# Filtrare la crew per il film "Following" e per regista Christopher Nolan
crew_following = crew[(crew['id'] == movie_id_following)]

# Filtrare la crew per il film "Memento"
crew_memento = crew[(crew['id'] == movie_id_memento)]

# Contare il numero di persone per ruolo per "Following"
crew_role_counts_following = crew_following['role'].value_counts()

# Contare il numero di persone per ruolo per "Memento"
crew_role_counts_memento = crew_memento['role'].value_counts()

# Filtrare gli attori per il film "Following"
actors_following = actors[actors['id'] == movie_id_following]
num_actors_following = len(actors_following)

# Filtrare gli attori per il film "Memento"
actors_memento = actors[actors['id'] == movie_id_memento]
num_actors_memento = len(actors_memento)

# Creazione dei grafici affiancati
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Grafico per "Following"
crew_role_counts_following.plot(kind='bar', color='skyblue', ax=ax1, label='Crew')
ax1.axhline(y=num_actors_following, color='red', linestyle='dashed', label=f'Actors: {num_actors_following}')
ax1.set_xlabel("Ruoli nella crew")
ax1.set_ylabel("Numero di persone")
ax1.set_title("Crew e attori in 'Following' (1998)")
ax1.legend()
ax1.tick_params(axis='x', rotation=90)

# Grafico per "Memento"
crew_role_counts_memento.plot(kind='bar', color='lightgreen', ax=ax2, label='Crew')
ax2.axhline(y=num_actors_memento, color='red', linestyle='dashed', label=f'Actors: {num_actors_memento}')
ax2.set_xlabel("Ruoli nella crew")
ax2.set_ylabel("Numero di persone")
ax2.set_title("Crew e attori in 'Memento' (2000)")
ax2.legend()
ax2.tick_params(axis='x', rotation=90)

# Visualizzare entrambi i grafici
plt.tight_layout()
plt.show()
```

Following, con il suo cast ridotto e una crew molto essenziale (conseguenza del low budget), dimostra il carattere fortemente indipendente del progetto, in cui Nolan stesso ricopriva più ruoli.
Memento, pur mantenendo una certa compattezza (specialmente nel numero di attori), segna un incremento notevole nella varietà e nella specializzazione dei ruoli tecnici. Questo è un segnale dell’aumento di risorse a disposizione e della volontà di Nolan di circondarsi di una squadra più strutturata e professionale.
 Questi dati non solo mostrano la crescita di Nolan tra il 1998 e il 2000, ma riflettono anche il passaggio da un cinema indipendente e sperimentale a una produzione capace di attrarre attenzione internazionale e riconoscimenti prestigiosi. Il successo di Memento è stato il vero trampolino di lancio, sia in termini di consensi critici, sia per l’accesso a una macchina produttiva più solida, ponendo le basi per la straordinaria carriera successiva.

# L’Ascesa con il Cinema Mainstream (2002-2008) e Il Successo Mondiale (2008-2014)



Dopo Memento, Nolan dirige Insomnia (2002), un thriller con Al Pacino e Robin Williams, che dimostra la sua capacità di lavorare con grandi star. Tuttavia, è nel 2005 che cambia il panorama cinematografico con Batman Begins, il primo capitolo della trilogia del Cavaliere Oscuro. Il film ridefinisce il genere dei cinecomic, portando realismo e profondità psicologica a un personaggio iconico. 
Nel 2008, Il Cavaliere Oscuro segna un punto di svolta. Il film, con Heath Ledger nei panni del Joker, ottiene incassi record e si aggiudica due Oscar. Questo successo permette a Nolan di sviluppare Inception (2010), un capolavoro di fantascienza che mescola sogno e realtà, vincitore di quattro premi Oscar.




```python
# Filtrare i film desiderati
films = ['Batman Begins', 'The Dark Knight', 'The Dark Knight Rises']
df_oscar_filtered = oscar_awards[oscar_awards['film'].isin(films)].copy()  # Creare una copia per evitare il SettingWithCopyWarning

# Creare una colonna per indicare se il film ha vinto
df_oscar_filtered.loc[:, 'vittoria'] = df_oscar_filtered['winner'].apply(lambda x: 1 if x else 0)

# Raggruppare per film e contare vittorie e candidature
df_oscar_counts = df_oscar_filtered.groupby('film').agg(
    vittorie=('vittoria', 'sum'),
    candidature=('vittoria', lambda x: len(x) - x.sum())
).reset_index()

# Plot del grafico
df_oscar_counts.set_index('film')[['vittorie', 'candidature']].plot(kind='bar', stacked=False, figsize=(10, 6))

# Aggiungere etichette e titolo
plt.title('Vittorie e Candidature agli Oscar per i Film Batman')
plt.xlabel('Film')
plt.ylabel('Numero')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrare il grafico
plt.show()
```

Batman Begins, primo capitolo della trilogia, ha ottenuto solo una candidatura agli Oscar senza portare a casa alcuna vittoria. Tuttavia, il film ha gettato le basi per il successo straordinario del suo seguito, Il Cavaliere Oscuro, che ha ricevuto un numero significativo di candidature e conquistato due prestigiose statuette. L’assenza de Il Cavaliere Oscuro - Il Ritorno nel grafico evidenzia come il terzo capitolo sia stato completamente ignorato dall’Academy, a differenza del secondo film, che ha segnato un punto di svolta nella storia del cinema.

Questa tendenza suggerisce un forte riconoscimento da parte degli Oscar per Il Cavaliere Oscuro, mentre il primo capitolo ha ottenuto un'attenzione limitata e il terzo è stato del tutto escluso. Resta ora da vedere se questa stessa dinamica si rifletta anche nei giudizi della critica e nel punteggio delle recensioni del pubblico.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrare i film desiderati
films = ['Batman Begins', 'The Dark Knight', 'The Dark Knight Rises']

# Calcolare il rating medio per ogni film
df_movies_filtered = movies[movies['name'].isin(films)]
rating_medio = df_movies_filtered.groupby('name')['rating'].mean()

# Calcolare il punteggio medio delle recensioni per ogni film
df_reviews_filtered = rotten_tomatoes[rotten_tomatoes['movie_title'].isin(films)]
review_score_medio = df_reviews_filtered.groupby('movie_title')['review_score'].mean()

# Unire i dati dei rating medi e dei punteggi delle recensioni
df_combined = pd.DataFrame({
    'Rating Medio Critica': rating_medio*2,  # Moltiplicato per 2 per adattarlo alla scala delle recensioni
    'Review Score Medio Pubblico': review_score_medio
}).reset_index()

# Impostare la larghezza delle barre
bar_width = 0.35
index = range(len(df_combined))

# Creare il grafico a barre
plt.figure(figsize=(10, 6))

# Tracciare le barre per rating medio e review score medio
plt.bar(index, df_combined['Rating Medio Critica'], bar_width, label='Rating Medio Critica', color='blue')
plt.bar([i + bar_width for i in index], df_combined['Review Score Medio Pubblico'], bar_width, label='Review Score Medio Pubblico', color='red')

# Aggiungere etichette e titolo
plt.title('Rating Medio e Review Score Medio per i Film Batman')
plt.xlabel('Film')
plt.ylabel('Punteggio')
plt.xticks([i + bar_width / 2 for i in index], df_combined['index'], rotation=45)
plt.legend()

# Mostrare il grafico
plt.tight_layout()
plt.show()
```

Analizzando ora i punteggi della critica e il giudizio del pubblico, emerge un quadro interessante. Batman Begins si attesta su valori compresi tra il 7 e l’8, segno di un’accoglienza positiva, seppur senza raggiungere l’eccellenza, soprattutto durante la notte degli Oscar. Il vero picco qualitativo si registra con Il Cavaliere Oscuro, che ottiene punteggi vicini all’8.5, riflettendo l’impatto straordinario che il film ha avuto sia sul pubblico che sulla critica. Infine, Il Cavaliere Oscuro - Il Ritorno riporta valori simili a quelli del primo capitolo, oscillando tra il 7 e l’8, a conferma di un solido apprezzamento, ma senza il clamore del film precedente.  

Questi dati suggeriscono che, nonostante la scarsa considerazione dell’Academy per il primo e il terzo film, entrambi sono stati ben accolti dal pubblico e dalla critica. Il secondo capitolo, invece, si distingue nettamente come il punto più alto della trilogia in termini di riconoscimento e impatto culturale. Questo dimostra come il valore di un film non si misuri esclusivamente con i premi vinti, ma anche con l'affetto e l'ammirazione che riesce a suscitare nel tempo.

```python
# Identificare i film di Batman
batman_movies = movies[movies['name'].str.contains("Batman", case=False, na=False)]

# Identificare i film di Batman diretti da Christopher Nolan
nolan_movies = crew[(crew['name'] == "Christopher Nolan") & (crew['role'] == "Director")]
nolan_batman_movies = batman_movies[batman_movies['id'].isin(nolan_movies['id'])]

# Identificare i film di Batman NON diretti da Christopher Nolan
non_nolan_batman_movies = batman_movies[~batman_movies['id'].isin(nolan_movies['id'])]

# Ottenere le recensioni per i film di Nolan e non-Nolan
nolan_reviews = rotten_tomatoes[rotten_tomatoes['movie_title'].isin(nolan_batman_movies['name'])].copy()
non_nolan_reviews = rotten_tomatoes[rotten_tomatoes['movie_title'].isin(non_nolan_batman_movies['name'])].copy()

# Convertire i punteggi in valori numerici se necessario
nolan_reviews.loc[:, 'review_score'] = pd.to_numeric(nolan_reviews['review_score'], errors='coerce')
non_nolan_reviews.loc[:, 'review_score'] = pd.to_numeric(non_nolan_reviews['review_score'], errors='coerce')

# Calcolare la media dei punteggi
nolan_avg_score = nolan_reviews['review_score'].mean()
non_nolan_avg_score = non_nolan_reviews['review_score'].mean()

# Stampare i risultati
print(f"Punteggio medio per i film di Batman diretti da Nolan: {nolan_avg_score:.2f}")
print(f"Punteggio medio per i film di Batman NON diretti da Nolan: {non_nolan_avg_score:.2f}")

# Creare un grafico a barre
plt.figure(figsize=(8, 5))
plt.bar(["Batman di Nolan", "Batman non di Nolan"], [nolan_avg_score, non_nolan_avg_score], color=["blue", "red"])
plt.xlabel("Categoria")
plt.ylabel("Punteggio medio")
plt.title("Confronto tra i punteggi dei film di Batman")
plt.ylim(0, 10)
plt.show()
```

Questi dati ci confermano come la trilogia di Batman di Christopher Nolan si distingue nettamente rispetto agli altri adattamenti cinematografici, confermando il loro impatto sia critico che popolare. Anche se non tutti i capitoli della trilogia hanno ricevuto il medesimo riconoscimento in termini di premi, il consenso unanime della critica e del pubblico dimostra che il suo approccio ha lasciato un segno indelebile nella storia del cinema.

Dopo l'enorme successo de Il Cavaliere Oscuro, Nolan si concentra sulla creazione di un nuovo capolavoro: Inception.

```python
# Filtrare i film desiderati
films = ['Inception', 'The Dark Knight']

# Trova gli ID dei film
inception_movie = movies[movies['name'] == 'Inception']
dark_knight_movie = movies[movies['name'] == 'The Dark Knight']

inception_id = inception_movie['id'].values[0]
dark_knight_id = dark_knight_movie['id'].values[0]

# Filtrare i generi per i film "Inception" e "The Dark Knight"
inception_genres = genres[genres['id'] == inception_id]['genre'].value_counts()
dark_knight_genres = genres[genres['id'] == dark_knight_id]['genre'].value_counts()

# Creare il grafico a torta per "Inception"
plt.figure(figsize=(14, 6))

# Inception
plt.subplot(1, 2, 1)
plt.pie(inception_genres, labels=inception_genres.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Generi di "Inception"')

# The Dark Knight
plt.subplot(1, 2, 2)
plt.pie(dark_knight_genres, labels=dark_knight_genres.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Generi di "The Dark Knight"')

# Mostrare i grafici
plt.tight_layout()
plt.show()
```

Uno degli aspetti più affascinanti del cinema di Christopher Nolan è la coerenza del suo stile, indipendentemente dal genere affrontato. Il Cavaliere Oscuro e Inception sono due capolavori riconosciuti a livello globale, eppure, pur appartenendo a universi narrativi differenti, condividono un’impronta autoriale ben riconoscibile. Nolan non ha dovuto adattarsi ai gusti del pubblico o reinventarsi radicalmente per ottenere successo: ha semplicemente applicato la sua visione a storie diverse, mantenendo intatta la sua cifra stilistica.  

Analizzando i temi dominanti nei due film, Il Cavaliere Oscuro si distribuisce equamente tra azione (25%), crime (25%), thriller (25%) e dramma (25%), risultando in un’opera cupa, complessa e moralmente ambigua, che trascende il genere supereroistico per diventare un crime movie di altissimo livello. D’altro canto, Inception sposta l’attenzione verso una miscela di avventura (33,3%), science fiction (33,3%) e azione (33,3%), mantenendo comunque la componente di tensione e intrigo che caratterizza tutta la sua filmografia.  

Questo dimostra come Nolan abbia saputo spaziare tra generi apparentemente diversi senza perdere la propria identità. Che si tratti di un cinecomic iperrealista o di un thriller fantascientifico basato sui sogni, il suo talento nel costruire trame intricate, nel giocare con la percezione della realtà e nel dare profondità ai personaggi rimane una costante, consolidando il suo status di autore tra i più influenti della sua generazione.

SPERIMENTA CON SCIENCE FICTION

```python
# Uniamo i dati dei film con i generi utilizzando l'id
movies_with_genres = pd.merge(movies, genres, left_on='id', right_on='id', how='inner')

# Filtriamo solo i film di genere 'Science Fiction'
science_fiction_movies = movies_with_genres[movies_with_genres['genre'] == 'Science Fiction']

# Filtra i film che sono usciti dal 2014 in poi
science_fiction_movies = science_fiction_movies[(science_fiction_movies['date'] >= 2014) & (science_fiction_movies['date'] <= 2020) ]

# Calcoliamo il numero di film di Science Fiction per anno
movies_count_by_year = science_fiction_movies.groupby('date').size()

# Impostiamo il grafico
plt.figure(figsize=(12, 6))

# Grafico del numero di film Science Fiction per anno
movies_count_by_year.plot(kind='line', marker='o', color='g', linestyle='-', linewidth=2, markersize=6)

# Aggiungi il titolo e le etichette
plt.title('Numero di Film Science Fiction dal 2014', fontsize=16)
plt.xlabel('Anno', fontsize=14)
plt.ylabel('Numero di Film', fontsize=14)
plt.xticks(range(2014, 2021), rotation=45, ha='right', fontsize=12)  # Etichette per gli anni dal 2014 in poi
plt.yticks(fontsize=12)
plt.grid(True)

# Ottimizza il layout
plt.tight_layout()
plt.show()
```
